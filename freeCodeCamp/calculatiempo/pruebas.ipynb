# Time calculator

Write a function named add_time that takes in two required parameters  
and one optional parameter:  
   - a start time in the 12-hour clock format(ending in AM or PM)  
   - a duration time that indicates the number of hours and minutes  
   - (optional) a starting day of the week, case insensitive  

The function should add the duration time to the start time and  
return the result.

If the result will be the next day, it should show(next day) after the time.  
If the result will be more than one day later, it should show(n days later)  
after the time, where "n" is the number of days later.  

If the function is given the optional starting day of the week parameter,  
then the output should display the day of the week of the result.  
The day of the week in the output should appear after the time and before  
the number of days later.

```
add_time("3:00 PM", "3:10")
# Returns: 6:10 PM

add_time("11:30 AM", "2:32", "Monday")
# Returns: 2:02 PM, Monday

add_time("11:43 AM", "00:20")
# Returns: 12:03 PM

add_time("10:10 PM", "3:30")
# Returns: 1:40 AM (next day)

add_time("11:43 PM", "24:20", "tueSday")
# Returns: 12:03 AM, Thursday (2 days later)

add_time("6:30 PM", "205:12")
# Returns: 7:42 AM (9 days later)
```

Y ahora vamos a crear la función

````python
def add_time(start, duration, input_day=None):

    start_list = start.split()  # ['3:00','PM']
    start_list[0] = start_list[0].split(':')  # [['3','00'],'PM']

    start_mins = int(start_list[0][0])*60 + int(start_list[0][1])  # 180

    if start_list[1] == 'PM':
        start_mins += 720  # 900

    dur_list = duration.split(':')  # ['23','10']
    dur_mins = int(dur_list[0])*60 + int(dur_list[1])  # 1390

    dur_total = (start_mins + dur_mins)  # 2290
    hours = dur_total // 60  # 38
    minutes = dur_total - hours * 60  # 10

    if (dur_total//720) % 2 == 0:
        am_pm = 'AM'
    else:
        am_pm = 'PM'  # 'PM'

    # para saber cuántas veces recorre el ciclo entero
    days = hours//24  # 1

    hour = hours - (hours//12)*12  # 2 (+ PM)
    if hour == 0:
        hour = 12

    #------------------WEEK DAYS------------------------

    week_list = ['Monday', 'Tuesday', 'Wednesday',
                'Thursday', 'Friday', 'Saturday', 'Sunday']

    if input_day:
        # ahora nos ubicamos en el index del día que se haya añadido como parámetro
        cod_sem = week_list.index(str(input_day).capitalize())  # Saturday = 5

        # para recorrer el bucle de días y que pase del 6 al 0
        counter = days
        while counter > 0:
            if cod_sem == 6:
                cod_sem = 0
            else:
                cod_sem += 1
            counter -= 1
        if days > 1:
            print(
                f'{hour}:{minutes:02d} {am_pm}, {week_list[cod_sem]} ({days} days later)')
        elif days == 1:
            print(
                f'{hour}:{minutes:02d} {am_pm}, {week_list[cod_sem]} (next day)')
        else:
            print(f'{hour}:{minutes:02d} {am_pm}, {week_list[cod_sem]}')

    else:
        if days > 1:
            print(f'{hour}:{minutes:02d} {am_pm} ({days} days later)')
        elif days == 1:
            print(f'{hour}:{minutes:02d} {am_pm} (next day)')
        else:
            print(f'{hour}:{minutes:02d} {am_pm}')
        


add_time("3:00 PM", "3:10")
# Returns: 6:10 PM

add_time("11:30 AM", "2:32", "Monday")
# Returns: 2:02 PM, Monday

add_time("11:43 AM", "00:20")
# Returns: 12:03 PM

add_time("10:10 PM", "3:30")
# Returns: 1:40 AM (next day)

add_time("11:43 PM", "24:20", "tueSday")
# Returns: 12:03 AM, Thursday (2 days later)

add_time("6:30 PM", "205:12")
# Returns: 7:42 AM (9 days later)
````
